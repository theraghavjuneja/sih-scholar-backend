```python
import time
from bs4 import BeautifulSoup
from langchain_community.document_loaders import AsyncHtmlLoader

url = "https://scholar.google.com/citations?hl=en&user=75x4pdcAAAAJ"
loader = AsyncHtmlLoader(url)

# Function to extract <div> tags from HTML content
def extract_div_tags(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    return soup.find_all('div')

# Measure time for lazy_load()
start_time = time.time()

# Iterate through the documents yielded by lazy_load
for document in loader.lazy_load():
    html_content = str(document) # or however the text content is accessed
    div_tags = extract_div_tags(html_content)
    for div in div_tags:
        print(div)

end_time = time.time()
print(f"Time taken with lazy_load(): {end_time - start_time} seconds")
```
